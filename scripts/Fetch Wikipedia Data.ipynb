```python
import torch
import numpy as np
import pandas as pd
import wikipedia
import re
```

```python
# Specify the title of the Wikipedia page
wiki = wikipedia.page('Samothrace temple complex')
# Extract the plain text content of the page
text = wiki.content
text
```

```python
# replace \n with space
text = text.replace('\n', ' ')
text = re.sub(r'==.*?==+', '', text)
# replace multiple spaces with single space
text = ' '.join(text.split())
# remove text between brackets
text = re.sub(r'\([^)]*\)', '', text)
text = re.sub(r'\[[^)]*\]', '', text)
```
